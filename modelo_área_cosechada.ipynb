```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import pickle
```

```python
# Leer tabla
datos = pd.read_excel("C:/Users/Lenovo/Desktop/Talento Tech/Proyecto/base_agrícola_2019 _2023.xlsx")
datos.head()
```

```python
# Verificar si hay cadenas vacías ("") o solo espacios en blanco
vacios = datos.applymap(lambda x: isinstance(x, str) and x.strip() == "")
vacios.sum()
```

```python
# Llenar datos vacio
datos = datos.applymap(lambda x: 2020 if isinstance(x, str) and x.strip() == "" else x)
datos.head()
```

```python
# Filtrar columnas relevantes
datos_relevantes = datos[['Cultivo','Área sembrada (ha)','Área cosechada (ha)','Departamento']]
datos_relevantes.head()
```

```python
# Cultivos con mayor área sembrada
cultivos = datos.groupby("Cultivo")[['Área sembrada (ha)', 'Área cosechada (ha)']].sum()
cultivos = cultivos.sort_values(by='Área sembrada (ha)', ascending=False)
cultivos.head(30)
```

```python
# Nueva tabla con cultivos seleccionados
cultivos_selec = datos_relevantes[datos_relevantes['Cultivo'].isin(['Café','Arroz','Palma de aceite','Maíz', 'Caña', 'Plátano', 'Cacao','Yuca','Papa','Aguacate','Frijol','Banano','Soya','Naranja','Mango','Limón','Ñame','Arveja','Piña','Coco','Tomate','Mandarina','Mora','Ahuyama','Fique','Cebolla de bulbo','Patilla','Chontaduro','Cebolla de rama'])]
cultivos_selec.head()
```

```python
# Aplicar One-Hot Encoding 
datos_finales_selec= pd.get_dummies(cultivos_selec, columns=['Cultivo','Departamento'],dtype=int)
datos_finales_selec.head()
```

```python
# Relación entre area sembrada y area cosechada : cuanto de esa area espero cosechar. 
# Preprocesamiento de datos para el modelo
X = datos_finales_selec.drop('Área cosechada (ha)', axis=1) 
y = datos_finales_selec['Área cosechada (ha)']

# Dividir registros en segmentos de entrenamiento y prueba
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.2, random_state=69)

# Entrenar el modelo
modelo = RandomForestRegressor()  # Crear una instancia del modelo
modelo.fit(X_entrenamiento, y_entrenamiento)

# Realizar predicciones
y_predicciones = modelo.predict(X_prueba)
print(y_predicciones)

# Calcular el r2
r2 = r2_score(y_prueba, y_predicciones)
print("R2:", r2)

# Realizar validación cruzada
scores = cross_val_score(modelo, X, y, cv=5)  # cv=5 para 5 pliegues
print("Puntajes de validación cruzada:", scores)
print("Media de R^2 de validación cruzada:", np.mean(scores))
```

```python
# Gráfico de dispersión de los valores reales vs las predicciones
plt.scatter(y_prueba, y_predicciones, color='blue', label='Predicciones')
plt.plot(y_prueba, y_prueba, color='red', label='Referencia Perfecta')  # Línea de referencia perfecta
plt.xlabel('Valores Reales')  # Asegúrate de usar mayúscula si así lo prefieres
plt.ylabel('Predicciones')
plt.title('Valores Reales vs Predicciones')  # Agregar un título al gráfico
plt.legend()
plt.grid(True)  
plt.show()
```
